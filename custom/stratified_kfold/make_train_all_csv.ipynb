```python
import os
import json
import numpy as np
import pandas as pd
```

```python
train_all_json = "/opt/ml/segmentation/input/data/train_all.json"
```

```python
with open(train_all_json, "r", encoding="utf8") as outfile:
    json_data = json.load(outfile)

images = json_data["images"]
categories = json_data["categories"]
annotations = json_data["annotations"]
```

```python
annotations[0]
```

```python
class_dict = {0 : 'Backgroud'}

for category in categories:
    class_dict[category['id']] = category['name']

class_dict
```

```python
image_id, class_name, class_id, segmentation = [], [], [], []

for annotation in annotations:
    image_id.append(annotation['image_id'])
    class_name.append(class_dict[annotation['category_id']])
    class_id.append(annotation['category_id'])
    segmentation.append(annotation['segmentation'])
```

```python
raw_data = {
    'image_id' : image_id,
    'class_name' : class_name,
    'class_id' : class_id,
    'segmentation' : segmentation   
}
```

```python
df = pd.DataFrame(raw_data)
```

```python
df.head()
```

```python
df.to_csv('../input/data/train_all.csv', sep=',', na_rep='NaN')
```
